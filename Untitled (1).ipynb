```r
library(tidyverse)
```

```r
df = read_csv(
    'https://raw.githubusercontent.com/fivethirtyeight/data/master/bechdel/movies.csv')
```

```r
clean = df%>%
filter(!grepl('#N/A', `intgross_2013$`))%>%
filter(!grepl('#N/A', `domgross_2013$`))%>%
mutate(adjdom = `domgross_2013$`, adjint = `intgross_2013$`)%>%
group_by(year)%>%
mutate(adjdom = as.numeric(adjdom)) %>%
mutate(adjint = as.numeric(adjint))%>%
mutate(wwgross = adjdom + adjint) %>%
mutate(roi = wwgross/budget)%>%
select(year, title, binary, budget, adjdom, adjint, wwgross, roi) %>%
arrange(year)
clean
```

```r
new = clean %>%
mutate(Decade = round(year/10)*10)%>%
group_by(Decade)%>%
mutate(Films = n())

new
```
